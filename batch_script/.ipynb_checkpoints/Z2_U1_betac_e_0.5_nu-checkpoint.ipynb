```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import sys
import os
import math
from statsmodels.graphics.tsaplots import plot_acf
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf
import scipy.integrate as integrate
import random
import h5py
from scipy.optimize import fsolve
from scipy.optimize import bisect


plt.rc('text',usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{bm}')
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.serif'] = 'Computer Modern'
plt.rcParams['axes.linewidth']  = 3.0
plt.rcParams['axes.labelsize']  = 30
plt.rcParams.update({'font.size': 22})
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.rcParams['xtick.major.size'] = 4
plt.rcParams['ytick.major.size'] = 4
plt.rcParams['xtick.minor.size'] = 3
plt.rcParams['ytick.minor.size'] = 3
plt.rcParams['legend.fontsize']  = 22
plt.rcParams['legend.frameon']  = False
plt.rcParams["legend.fancybox"] = False
plt.rcParams["legend.shadow"] = False
plt.rcParams["legend.framealpha"] = 1
plt.rcParams["axes.facecolor"] = 'white'
plt.rcParams["axes.edgecolor"] = 'black'

plt.rcParams['figure.figsize'] = 8, 8

plt.rcParams['xtick.major.width'] = 1
plt.rcParams['ytick.major.width'] = 1
plt.rcParams['xtick.minor.width'] = 0
plt.rcParams['ytick.minor.width'] = 0

folder_Fig=("/Users/ilaria/Desktop/Multicomponents_SC/Output_3C/Report/Figures_e0.5")

bbox_props = dict(boxstyle="round", fc="w", ec="0.5", alpha=0.9)


################################################

e=0.5
h=5.4

nu=np.array([0, 0.1, 0.2])

###############################################

################ Z2 transition ##################

betac_z2=np.zeros((len(nu)))
err_betac_z2=np.zeros((len(nu)))


###############################################
#                                             #
#                 nu=0;                       #
#                                             #
###############################################

n=0
beta_low=0.2245
beta_high=0.229
nbeta=32
L=np.array([8,10,12])

if( (nu[n]).is_integer()): nu_n=int(nu[n])

folder_out=("/Users/ilaria/Desktop/MultiComponents_SC/Output_3C/e_%s/nu_%s/h_%s" %(e, nu_n, h))

fig, ax1 = plt.subplots(nrows=1, figsize=(6,6))
#ax1.set_title(r"$h=%s$; $e=%s$; $\nu=%s$" %(h, e, nu[n]))
ax1.set_xlabel(r"$\beta$")
ax1.set_ylabel(r"$U$")

color=iter(plt.cm.rainbow(np.linspace(0,1,len(L)+1)))

U_pol=[]
for l in range(len(L)):
    c_m=next(color)
    BASEDIR=("%s/L%d_a0_b1_eta1_e%s_h%s_nu%s_bmin%s_bmax%s"  %(folder_out, L[l], e,  h, nu_n, beta_low, beta_high))
    data_U= np.loadtxt("%s/Binder_cumulant.txt" %(BASEDIR))
    beta= data_U[0]
    U_mean= data_U[1]
    U_err= data_U[2] 
    U_pol.append(np.poly1d(np.polyfit(beta, U_mean, 4 )))
    #ax1.plot(beta, U_pol[l](beta), '-', c=c_m)
    ax1.plot(beta, U_mean, '-', c=c_m)
    ax1.errorbar(beta, U_mean, yerr=U_err, capsize=2, c=c_m, label="L=%s" %L[l])

ax1.legend(loc="best")
#ax1.annotate(r"$\nu=%s$ " %nu_n + "\n"+ "$\,\,h=%s$" %h + "\n"+ "$\,\,e=%s$" %e, (0.15, 0.15), xycoords='axes fraction',  ha="center", va="center", bbox=bbox_props)
ax1.annotate(r"$\nu=%s$" %nu_n, (0.15, 0.15), xycoords='axes fraction',  ha="center", va="center", bbox=bbox_props)
ax1.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
plt.tight_layout()
plt.savefig("%s/Binder_cumulant_h%s_nu%s.png" %(folder_Fig, h, nu_n))

betac_cross=[]
for l1 in range(len(L)):
    for l2 in range(l1+1, len(L)):
        Diff = U_pol[l1] -  U_pol[l2]
        #Return the roots of the (non-linear) equations defined by func(x) = 0 given a starting estimate
        betac_cross.append(bisect(Diff, beta[0], beta[-1]))
        
betac_z2[n]=np.mean(betac_cross)   
err_betac_z2[n]=np.sqrt(np.var(betac_cross)/(len(betac_cross)-1))

    
###############################################
#                                             #
#                 nu=0.1;                     #
#                                             #
###############################################
n=1
beta_low=0.2165
beta_high=0.2195
nbeta=32
L=np.array([8,10,12, 16, 20])

nu_n=nu[n]
if( (nu[n]).is_integer()): nu_n=int(nu[n])
folder_out=("/Users/ilaria/Desktop/MultiComponents_SC/Output_3C/e_%s/nu_%s/h_%s" %(e, nu_n, h))

fig, ax1 = plt.subplots(nrows=1, figsize=(6,6))
#ax1.set_title(r"$h=%s$; $e=%s$; $\nu=%s$" %(h, e, nu[n]))
ax1.set_xlabel(r"$\beta$")
ax1.set_ylabel(r"$U$")

color=iter(plt.cm.rainbow(np.linspace(0,1,len(L)+1)))
U_pol=[]

for l in range(len(L)):
    c_m=next(color)
    BASEDIR=("%s/L%d_a0_b1_eta1_e%s_h%s_nu%s_bmin%s_bmax%s"  %(folder_out, L[l], e,  h, nu_n, beta_low, beta_high))
    data_U= np.loadtxt("%s/Binder_cumulant.txt" %(BASEDIR))
    beta= data_U[0]
    U_mean= data_U[1]
    U_err= data_U[2] 
    U_pol.append(np.poly1d(np.polyfit(beta, U_mean, 4 )))
    #ax1.plot(beta, U_pol[l](beta), '-', c=c_m)
    ax1.plot(beta, U_mean, '-', c=c_m)
    ax1.errorbar(beta, U_mean, yerr=U_err, capsize=2, c=c_m, label="L=%s" %L[l])

ax1.legend(loc="best")
#ax1.annotate(r"$\nu=%s$" %nu_n + "\n"+ "$\,h=%s$" %h + "\n"+ "$\,e=%s$" %e, (0.15, 0.15), xycoords='axes fraction',  ha="center", va="center", bbox=bbox_props)
ax1.annotate(r"$\nu=%s$" %nu_n, (0.15, 0.15), xycoords='axes fraction',  ha="center", va="center", bbox=bbox_props)
ax1.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
plt.tight_layout()
plt.savefig("%s/Binder_cumulant_h%s_nu%s.png" %(folder_Fig, h, nu_n))

betac_cross=[]
for l1 in range(len(L)):
    for l2 in range(l1+1, len(L)):
        Diff = U_pol[l1] -  U_pol[l2]
        print(bisect(Diff, beta[0], beta[-1]))
        betac_cross.append(bisect(Diff, beta[0], beta[-1]))
        
betac_z2[n]=np.mean(betac_cross)
err_betac_z2[n]=np.sqrt(np.var(betac_cross)/(len(betac_cross)-1))

    
###############################################
#                                             #
#                 nu=0.2;                     #
#                                             #
###############################################
n=2
beta_low=0.208
beta_high=0.212
nbeta=32
L=np.array([8,10,12, 16, 20])

nu_n=nu[n]
if( (nu[n]).is_integer()): nu_n=int(nu[n])
folder_out=("/Users/ilaria/Desktop/MultiComponents_SC/Output_3C/e_%s/nu_%s/h_%s" %(e, nu_n, h))

fig, ax1 = plt.subplots(nrows=1, figsize=(6,6))
#ax1.set_title(r"$h=%s$; $e=%s$; $\nu=%s$" %(h, e, nu[n]))
ax1.set_xlabel(r"$\beta$")
ax1.set_ylabel(r"$U$")

color=iter(plt.cm.rainbow(np.linspace(0,1,len(L)+1)))

U_pol=[]

for l in range(len(L)):
    c_m=next(color)
    BASEDIR=("%s/L%d_a0_b1_eta1_e%s_h%s_nu%s_bmin%s_bmax%s"  %(folder_out, L[l], e,  h, nu_n, beta_low, beta_high))
    data_U= np.loadtxt("%s/Binder_cumulant.txt" %(BASEDIR))
    beta= data_U[0]
    U_mean= data_U[1]
    U_err= data_U[2] 
    U_pol.append(np.poly1d(np.polyfit(beta, U_mean, 4 )))
    #ax1.plot(beta, U_pol[l](beta), '-', c=c_m)
    ax1.plot(beta, U_mean, '-', c=c_m)
    ax1.errorbar(beta, U_mean, yerr=U_err, capsize=2, c=c_m, label="L=%s" %L[l])
ax1.legend(loc="best")
#ax1.annotate(r"$\nu=%s$" %nu_n + "\n"+ "$\,h=%s$" %h + "\n"+ "$\,e=%s$" %e, (0.15, 0.15), xycoords='axes fraction',  ha="center", va="center", bbox=bbox_props)
ax1.annotate(r"$\nu=%s$" %nu_n, (0.15, 0.15), xycoords='axes fraction',  ha="center", va="center", bbox=bbox_props)
ax1.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
plt.tight_layout()
plt.savefig("%s/Binder_cumulant_h%s_nu%s.png" %(folder_Fig, h, nu_n))

betac_cross=[]
for l1 in range(len(L)):
    for l2 in range(l1+1, len(L)):
        Diff = U_pol[l1] -  U_pol[l2]
        print(bisect(Diff, beta[0], beta[-1]))
        #Return the roots of the (non-linear) equations defined by func(x) = 0 given a starting estimate
        betac_cross.append(bisect(Diff, beta[0], beta[-1]))
        
betac_z2[n]=np.mean(betac_cross)
err_betac_z2[n]=np.sqrt(np.var(betac_cross)/(len(betac_cross)-1))

    
```
