```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import sys
import os
import math
from statsmodels.graphics.tsaplots import plot_acf
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf
import scipy.integrate as integrate
import random
import h5py
from scipy.optimize import fsolve
from scipy.optimize import bisect

plt.rc('text',usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{bm}')
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.serif'] = 'Computer Modern'
plt.rcParams['axes.linewidth']  = 3.0
plt.rcParams['axes.labelsize']  = 30
plt.rcParams.update({'font.size': 22})
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.rcParams['xtick.major.size'] = 4
plt.rcParams['ytick.major.size'] = 4
plt.rcParams['xtick.minor.size'] = 3
plt.rcParams['ytick.minor.size'] = 3
plt.rcParams['legend.fontsize']  = 22
plt.rcParams['legend.frameon']  = True
plt.rcParams["legend.fancybox"] = True
plt.rcParams["legend.shadow"] = True
plt.rcParams["legend.framealpha"] = 1
plt.rcParams["axes.facecolor"] = 'white'
plt.rcParams["axes.edgecolor"] = 'black'

plt.rcParams['figure.figsize'] = 8, 8

plt.rcParams['xtick.major.width'] = 1
plt.rcParams['ytick.major.width'] = 1
plt.rcParams['xtick.minor.width'] = 0
plt.rcParams['ytick.minor.width'] = 0

################################################

e=0
h=0.2
nu=0
beta_low=10.5
beta_high=11.75

###############################################


color=iter(plt.cm.rainbow(np.linspace(0,1,6)))

HOMEDIR=("/Users/ilaria/Desktop/MultiComponents_SC/Output_3C/e_%s/nu_%s/h_%s" %(e, nu, h))

L=np.array([8, 10, 12])

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=2, sharex=True, figsize=(10,10))
fig.suptitle(r"$\nu=%s$" %nu)
ax1[0].set_ylabel(r"$U$")
ax2[0].set_ylabel(r"$L \Upsilon$")
ax1[1].set_ylabel(r"$L \rho$")
ax2[1].set_ylabel(r"$|\Psi_0|$")
ax2[1].set_xlabel(r"$\beta$")
ax1[0].set_yscale('linear')
ax2[0].set_yscale('linear')
ax2[0].set_xlabel(r"$\beta$")
ax1[0].set_xlim((10.75, 11.5))
ax2[0].set_xlim((10.75, 11.5))
ax1[1].set_xlim((10.75, 11.5))
ax2[1].set_xlim((10.75, 11.5))
ax1[0].grid()
ax2[0].grid()

for l in range(len(L)):
    BASEDIR=("%s/L%d_a0_b1_eta1_e%s_h%s_nu%s_bmin%s_bmax%s"  %(HOMEDIR, L[l], e,  h, nu, beta_low, beta_high))
    file_Js_sum=("%s/Helicity_modulus_sum.txt" %(BASEDIR))
    beta, Js_sum, Js_sum_err= np.loadtxt(file_Js_sum, usecols=(0,1,2), unpack=True)
    
    file_Binder=("%s/Binder_cumulant.txt" %(BASEDIR))
    data_U= np.loadtxt(file_Binder)
    beta=data_U[0,:]
    U=data_U[1,:]
    U_err=data_U[2,:]

    file_DS=("%s/Dual_Stiffness.txt" %(BASEDIR))
    data_DS= np.loadtxt(file_DS)
    beta=data_DS[0,:]
    DS=data_DS[1,:]
    DS_err=data_DS[2,:]
    
    file_Psi=("%s/Psi_Density.txt" %(BASEDIR))
    data_Psi= np.loadtxt(file_Psi)
    beta=data_Psi[0,:]
    Psi1=data_Psi[1,:]
    Psi1_err=data_Psi[2,:]
    Psi2=data_Psi[3,:]
    Psi2_err=data_Psi[4,:]   
    Psi2=data_Psi[5,:]
    Psi2_err=data_Psi[6,:]  
    

    ax1[0].errorbar(beta, U, yerr=U_err, fmt='o-', label="%d" %L[l])
    ax2[0].errorbar(beta, L[l]*float(h)*Js_sum, yerr=Js_sum_err, fmt='o-')
    ax1[1].errorbar(beta, L[l]*float(h)*DS, yerr=DS_err, fmt='o-')
    ax2[1].errorbar(beta, Psi1, yerr=Psi1_err, fmt='o-')
    
    #ax1.legend(title="L", loc='best')
    lines = []
    labels = []
    for ax in fig.axes:
        axLine, axLabel = ax.get_legend_handles_labels()
        lines.extend(axLine)
        labels.extend(axLabel)

fig.tight_layout()
fig.subplots_adjust(bottom=0.2, top=0.92)
fig.legend(lines, labels, loc = 'lower center', title="L", ncol=len(L))    
fig.savefig("%s/Critical_points_bmin%s_bmax%s.png" %(HOMEDIR, beta_low, beta_high))
plt.show()
```
